### Problem of Interest

We wish to understand the severity of a collision based on a variety of different aspects. We will focus on data that includes not only severity of collision and injuries, but also data that we feel could and would contribute to a collision. This information could be used in a variety of different aspects of the business world including but not limiting to: Insurance, Healthcare, Emergency Preparedness and Personel, and Motor Vehicle sales and repair. 

#### How the problem ties to the various business models

__Insurance__ : The severity of collision would indicate costs associated with such a collision. By being able to predict the severity of a collision based on some aspects of the surrounding environment, an insurance company could make better rates: increasing rates for drivers in poor conditions and decreasing rates for drivers under good conditions. You would only be able to use static variables such as location, climate, car type, etc. for this prediction as daily changing variables (except maybe averages) could not be included in a six month contract.

__Healthcare__ : The severity of collisions would be very beneficial to healthcare. Knowing the conditions at which severe injury occurs would allow a hospital or emergency room to better plan both the need for supplies and the need to schedule personel at the appropriate times. This would allow the hospital to run more efficiently and smoothly.

__Emergency Preparedness__ : The severity of collisions would allow cities, counties, and states to better plan their emergency personel including police presence, fire department size, and number of ambulances available at certain times. It would allow for long term planning making sure they have enough basic supplies and vehicles when static conditions, such as time of year, indicate that more severe accidents will occur as well as adjusting short term resources such as how many individuals are scheduled to be on call when easily changed variables, such as predicted rainfall, change.

__Motor Vehicle Sales__ : If static conditions indicate there will be more severe collisions in a certain area, this will allow for better planning of supply as more vehicles will be needed to replace the vehicles that are totalled in the collision. It will also allow for the better planning of what types of vehicles are needed. If static conditions indicate that there will be more severe collisions, after a time the general public will want vehicles that are better at protecting the passengers in said vehicle. 

__Motor Vehicle Repair__ : If static conditions indicate there will be less severe collisions then businesses that repair vehicles can resonably assume that the vehicles can be fixed or at least salvagable. This may indicate the need to hire more mechanics or expand garage space so that the business has the ability to handle a larger work load.

#### Machine Learning

Since we are trying to produce a way to classify which variables cause a more severe accident, it makes sense to use classification Machine Learning Algorithms such as K-nearest neighbors and decision trees. At the end we will calculate the Jaccard and f1 scores to get a sense of accuracy.

In order to do this, we must adjust our data to be all numerical data. We do this using One-Hot Encoding. Most data, split nicely. I include counts of data types that we will not have enough data for certain values.

```python


#Innattentionind has black values for when there were no data or when there wasn't a lack of inattention.
#We will do the analysis again for when we have information in the next section. 

#UNDERINFL was recorded in multiple ways, here we consolidate the data into the same values. 

dfml['UNDERINFL']=dfml['UNDERINFL'].replace(['0','1', 'N', 'Y', 'No', 'Yes'],['0','1', '0', '1', '0', '1'])



#We need to one hot encode the variables that are strings. 
dfml = pd.concat([dfml,pd.get_dummies(dfml['STATUS'])], axis=1)
dfml.drop(['STATUS'], axis = 1,inplace=True)
dfml = pd.concat([dfml,pd.get_dummies(dfml['ADDRTYPE'])], axis=1)
dfml.drop(['ADDRTYPE'], axis = 1,inplace=True)
dfml = pd.concat([dfml,pd.get_dummies(dfml['COLLISIONTYPE'])], axis=1)
dfml.drop(['COLLISIONTYPE'], axis = 1,inplace=True)
dfml = dfml.rename(columns={'Other': 'Other Angle'})
dfml = pd.concat([dfml,pd.get_dummies(dfml['WEATHER'])], axis=1)
dfml.drop(['WEATHER'], axis = 1,inplace=True)
dfml = dfml.rename(columns={'Other': 'Other Weather'})


#Converting weather types to another type of variable so we can add them.
dfml['Other Angle']=dfml['Other Angle'].astype('float')
dfml['Other Weather']=dfml['Other Weather'].astype('float')
dfml['Blowing Sand/Dirt']=dfml['Blowing Sand/Dirt'].astype('float')
dfml['Partly Cloudy']=dfml['Partly Cloudy'].astype('float')
dfml['Severe Crosswind']=dfml['Severe Crosswind'].astype('float')

#We drop the weather columns Other, Blowing Dirt/Sand, Partly Cloudy, and Severe Crosswind.


dfml.drop(['Unknown'], axis = 1,inplace=True)
dfml.drop(['Blowing Sand/Dirt'], axis = 1,inplace=True)
dfml.drop(['Partly Cloudy'], axis = 1,inplace=True)
dfml.drop(['Severe Crosswind'], axis = 1,inplace=True)

```

```python
#Do to the fact that some variables have the same strings such as unknown. We split the datacleaning. We continue the one hot encoding.

dfml = pd.concat([dfml,pd.get_dummies(dfml['LIGHTCOND'])], axis=1)
dfml.drop(['LIGHTCOND'], axis = 1,inplace=True)

#We drop the Dark-Unknown Lighting due to small sample size and the Unknown variable as we cannot know how to apply this.
dfml.drop(['Dark - Unknown Lighting'], axis = 1,inplace=True)
dfml.drop(['Unknown'], axis = 1,inplace=True)

#Now for the PEDROWNOTGRNT variable
# There are too many NaN values. These could be analyzed independent of the rest of the data

#dfml = pd.concat([dfml,pd.get_dummies(dfml['PEDROWNOTGRNT'])], axis=1)
#dfml.drop(['PEDROWNOTGRNT'], axis = 1,inplace=True)
#dfml = dfml.rename(columns={'Y': 'Failed Pedestrian Right of Way'})

dfml = pd.concat([dfml,pd.get_dummies(dfml['HITPARKEDCAR'])], axis=1)
dfml.drop(['HITPARKEDCAR'], axis = 1,inplace=True)
dfml = dfml.rename(columns={'Y': 'Hit Parked Car'})
dfml = dfml.rename(columns={'N': 'Did Not Hit Parked Car'})

dfml.head()
```

Now we train the model. We will do a variety of values of K. 

```python
# Modeling


# Best k
Ks=15
mean_acc=np.zeros((Ks-1))
std_acc=np.zeros((Ks-1))
ConfustionMx=[];
for n in range(1,Ks):
    
    #Train Model and Predict  
    kNN_model = KNeighborsClassifier(n_neighbors=n).fit(X_train,y_train)
    yhat = kNN_model.predict(X_test)
    
    
    mean_acc[n-1]=np.mean(yhat==y_test);
    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
mean_acc

```

As we can see, we can't really get much above a .74 rating so the best k would be the smallest k around that value. This would be k=8. I rerun K=8 to have that specific model. 

```python
k = 8
from sklearn.neighbors import KNeighborsClassifier
#Train Model and Predict  
kNN_model8 = KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
kNN_model8
```
